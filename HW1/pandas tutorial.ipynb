```python
import pandas as pd
import numpy as np
import random
```

```python
EMOTIONS = ["anger", "disgust", "fear", "joy", "sadness", "surprised", "neutral"]
index_points = list(np.arange(48, 69))

# containing points from 48 to 68 
each_emotion_points = [(int(random.random()*100), int(random.random()*100)) for i in range(48, 69)]
every_emotion_points = [[(int(random.random()*100), int(random.random()*100)) for i in range(48, 69)] for i in range(7)]
```

```python
df = pd.DataFrame()
for i in range(len(EMOTIONS)):
    df[EMOTIONS[i]] = pd.Series(every_emotion_points[i], index=index_points)
```
