```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import scipy.stats
```

```python
all = pd.read_csv('hr-employee-attrition-with-null.csv')
```

```python
all.loc[all["Attrition"] == "No", "Attrition"] = 0.0
all.loc[all["Attrition"] == "Yes", "Attrition"] = 1.0
```

```python
all = all.drop('Unnamed: 0', axis=1)
```

```python
encoded = ["BusinessTravel", "Department", "EducationField", "JobRole", "MaritalStatus", "Gender", "Over18", "OverTime"]
for col in encoded:
    all[col] = pd.Categorical(all[col]).codes
```

```python
all = all.drop(columns = "EmployeeNumber")
```

```python
attrition_yes = all[all["Attrition"] == 1]
attrition_no = all[all["Attrition"] == 0]

divided_yes = int(len(attrition_yes.index) * 0.90)
divided_no = int(len(attrition_no.index) * 0.90)

train_set = pd.DataFrame([], columns=all.columns, dtype=np.float32)
train_set = train_set.append(attrition_yes[:divided_yes], ignore_index=True)
train_set = train_set.append(attrition_no[:divided_no], ignore_index=True) 


test_set = pd.DataFrame([], columns=all.columns,dtype=np.float32)
test_set = test_set.append(attrition_yes[divided_yes:], ignore_index=True)
test_set = test_set.append(attrition_no[divided_no:], ignore_index=True)

```

# Histogram discretization (T4 & T5)

```python
bin_param = 40
for col in ["Age", "MonthlyIncome", "DistanceFromHome"]:
    #remove NaN values
    train_col_no_nan = train_set[~np.isnan(train_set[col])][col]
    
    # bin the data into 40 equally spaced bins 
    # hist is the count for each bin 
    # bin_edge is the edge values of the bins 
    hist, bin_edge = np.histogram(train_col_no_nan, bin_param)
    #print('Zero count percentage : ', list(hist).count(0) / bin_param)
    print('zero count : ', list(hist).count(0))

    plt.grid(True)
    plt.title(col)
    plt.ylabel('frequency')
    plt.xlabel('data')
    x = plt.fill_between(bin_edge.repeat(2)[1:-1],hist.repeat(2),facecolor='steelblue') 
    if (col == "Age"):
        plt.savefig("./T4_Age.png")
    elif (col == "MonthlyIncome"):
        plt.savefig("./T4_MonthlyIncome.png")
    elif (col == "DistanceFromHome"):
        plt.savefig("./T4_DistanceFromHome.png")
#     plt.show()
```

```python
y_true = np.array(test_set["Attrition"], dtype=np.int)
y_true
```

```python
effected_col = ["Age", "DailyRate", "DistanceFromHome", "EnvironmentSatisfaction", "OverTime", "RelationshipSatisfaction", "WorkLifeBalance"]
```

```python
def evaluate(y_true, y_pred):
    true_positive = 0
    false_positive = 0
    false_negative = 0
    true_negative = 0
    
    if (len(y_true) != len(y_pred)):
        print('Crash!!')
        return 
    
    for idx in range(len(y_true)):
        if y_true[idx] == 1 and y_pred[idx] == 1:
            true_positive += 1
        elif y_true[idx] == 1 and y_pred[idx] == 0:
            false_negative += 1
        elif y_true[idx] == 0 and y_pred[idx] == 1:
            false_positive += 1
        elif y_true[idx] == 0 and y_pred[idx] == 0:
            true_negative += 1
        else:
            print("Crash")
            return
    try: 
        precision = true_positive / (true_positive + false_positive)
        recall = true_positive / (true_positive + false_negative)
        f1_score = 2 * precision * recall / (precision + recall)
        accuracy = (true_positive + true_negative) / (true_positive + true_negative + false_positive + false_negative)
        print('Accuracy : ', accuracy)
        print('Precision : ', precision)
        print('Recall : ', recall)
        print('F1 score : ', f1_score)
    except ZeroDivisionError:
        print('True positive : ', true_positive)
        print('False positive : ', false_positive)
        print('False negative : ', false_negative)
        print('True negative : ', true_negative)
        print("Cannot solve according to your interested metrics")
```

```python
p_leave = (test_set[test_set["Attrition"] == 1]).shape[0] / test_set.shape[0]
p_stay = 1 - p_leave
```

```python
epsilon = 1e-10
y_pred = list()
for row_idx in range(test_set.shape[0]):
    isLeave = False 
    likelihood_H_x = math.log(p_leave) - math.log(p_stay) 
    for col in effected_col:
        feature_value = test_set.iloc[row_idx][col]
        p_x_given_leave = (test_set[(test_set["Attrition"] == 1) & (test_set[col] == feature_value)]).shape[0] / test_set[test_set["Attrition"] == 1].shape[0]
        p_x_given_stay = (test_set[(test_set["Attrition"] == 0) & (test_set[col] == feature_value)]).shape[0] / test_set[test_set["Attrition"] == 0].shape[0]
       
        if not (abs(p_x_given_leave - 0) < epsilon or abs(p_x_given_stay - 0) < epsilon):
            likelihood_H_x += math.log(p_x_given_leave, math.e)
            likelihood_H_x -= math.log(p_x_given_stay, math.e)
    if (likelihood_H_x > 0):
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
evaluate(y_true, y_pred)
```

```python
y_norm_pred = list()

yes_test_attrition = test_set[test_set["Attrition"] == 1]
no_test_attrition = test_set[test_set["Attrition"] == 0]

for row_idx in range(test_set.shape[0]):
    likelihood_H_x_norm = math.log(p_leave)- math.log(p_stay)
    
    for col in effected_col:
        isLeave = False
        yes_test_mean = yes_test_attrition[col].mean()
        yes_test_std = yes_test_attrition[col].std()
        no_test_mean = no_test_attrition[col].mean()
        no_test_std = no_test_attrition[col].std()
    
        feature_value = test_set.iloc[row_idx][col]

        yes_normal_value = scipy.stats.norm(yes_test_mean, yes_test_std).pdf(feature_value)
        no_normal_value = scipy.stats.norm(no_test_mean, no_test_std).pdf(feature_value)

        if not (abs(yes_normal_value - 0) < epsilon or abs(no_normal_value - 0) < epsilon or np.isnan(yes_normal_value) or np.isnan(no_normal_value)):
          
            likelihood_H_x_norm += math.log(yes_normal_value, math.e)
            likelihood_H_x_norm -= math.log(no_normal_value, math.e)
    
    if (likelihood_H_x_norm > 0):
        y_norm_pred.append(1)
    else:
        y_norm_pred.append(0)
```

```python
evaluate(y_true, y_norm_pred)
```
